```python
def regression_report(y_true,y_pred):
    print(f"""
    MSE:  {metrics.mean_squared_error(y_true,y_pred)}
    RMSE: {metrics.mean_squared_error(y_true,y_pred,squared=False)}
    MAE   {metrics.median_absolute_error(y_true,y_pred)}
    R2:   {metrics.r2_score(y_true,y_pred)}""")
```

```python
X = pd.DataFrame(california_housing_data["data"],
             columns=california_housing_data["feature_names"])
y = pd.Series(california_housing_data["target"],
             name=california_housing_data["target_names"][0])
```

```python
X = X.to_numpy()
y = y.to_numpy()
y=y.reshape(-1,1)
```

```python
# X = StandardScaler().fit_transform(X)
# y = StandardScaler().fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
X_train.shape,y_train.shape
```

```python
X_train_scaled = (X_train-X_train.mean(axis=0))/   \
                        X_train.std(axis=0)
```

```python
X_test_scaled = (X_test-X_train.mean(axis=0))/   \
                        X_train.std(axis=0)
```

```python
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
def poly_feature(X,degrees=3):
    return np.stack([X**degree 
               for degree in range(1,degrees+1)],axis=1)
```

```python
# X_train_scaled_poly = poly_feature(X_train_scaled,3)
# X_test_scaled_poly = poly_feature(X_test_scaled,3)

X_train_scaled_poly = PolynomialFeatures(3).fit_transform(X_train_scaled)
X_test_scaled_poly = PolynomialFeatures(3).fit_transform(X_test_scaled)

# X_train_scaled_poly = poly_feature(X_train_scaled,3)
# X_test_scaled_poly = poly_feature(X_test_scaled,3)
```

```python
def f_model(reqularizer="l2",
            loss='mean_squared_logarithmic_error',
            optimizer="SGD",
            epochs=100,
            batch_size=10,
            metrics=["mae",root_mean_squared_error,'mean_squared_logarithmic_error']):
     
    hp_learning_rate = hp.Choice('learning_rate', 
                                 values=[1e-2, 1e-3, 1e-4])
    hp_regularizer_alpha = hp.Choice('regularizer_alpha', 
                                     values=[1.0,0.1,0.01])
    
    model = Sequential([
    Dense(1,input_shape=X_train_scaled_poly[0].shape,
          activation="linear",
          kernel_initializer='ones',
          kernel_regularizer=tf.keras.regularizers.L2(l2=hp_regularizer_alpha))
    ])

    model.compile(loss=loss,
                  optimizer=tf.keras.optimizers.SGD(
                      learning_rate=hp_learning_rate),
                  metrics=metrics)
    return model
```

```python
def poly_feature_2d(X,degrees=3):
    return np.concatenate([X_train**degree 
                           for degree in range(1,degrees+1)],axis=1)
```

```python
X_train_poly = PolynomialFeatures(3).fit_transform(X_train)
X_test_poly = PolynomialFeatures(3).fit_transform(X_test)
# X_train_poly = poly_feature_2d(X_train,3)
# X_test_poly = poly_feature_2d(X_test,3)
```

```python
def create_and_fit_model(X,y,reqularizer="l2",
                         loss='mean_squared_logarithmic_error',
                         optimizer="SGD",
                         epochs=100,
                         batch_size=10,
                         metrics=["mae","mse",'mean_squared_logarithmic_error']):
    model = Sequential([
    Dense(1,input_dim=X.shape[1],
          activation="linear",
          kernel_initializer='ones',
          kernel_regularizer=reqularizer)
    ])

    model.compile(loss=loss,
                  optimizer=optimizer,
                  metrics=metrics)

    history=model.fit(X,
                      y,
                      batch_size=batch_size,
                      use_multiprocessing=True,
                      epochs=epochs,
                      verbose=1)
    plt.plot(history.history["loss"],title="loss")
    plt.plot(history.history["mse"],title="mse")
    return model,history
```

```python
grid_params = {"optimizer":["SGD",
                     "Adam","Adagrad","Adadelta"],
       "regularization":[l2(1),l2(0.1),
                         l2(0.01),l2(0.001)],
       "loss":["mean_squared_logarithmic_error","mae"],
       "batch_size":[1,20,50]}
```

```python
grid_list = [*itertools.product(*grid_params.values())]
```

```python
checkpoint_filepath = 'checkpoint'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
models_scores = []
for grid in tn(grid_list):
    model,history=create_and_fit_model(X_train_poly,y_train,
                              optimizer=grid[0],
                              reqularizer=grid[1],
                              loss = grid[2],
                              batch_size=grid[3],
                                      epochs=40)
    y_pred = model.predict(X_test_poly)
    rmse=metrics.mean_squared_error(y_test,y_pred,squared=False)
    models_scores.append((model,history,rmse))
```

```python
models_scores = pd.DataFrame(models_scores).sort_values(2)
```

```python
best_model = models_scores.iloc[0][0]
```

```python
regression_report(y_train,best_model.predict(X_train_poly))
```

```python
regression_report(y_test,best_model.predict(X_test_poly))
```

```python
y_pred_test = best_model.predict(X_test_poly)
```

```python
metrics.mean_absolute_percentage_error(y_test,y_pred_test)
```
