# data 1

```python
X_data = np.linspace(-1,1,101)
num_coef=10
coef =[-100,2,3,-3000,5,6,4,300,2,600]
y_data=0
for i in range(num_coef):
    y_data += coef[ i ]*np.power(X_data, i)
    y_data += np.random.randn(*X_data.shape )*100.5
```

```python
fig,ax = plt.subplots()
ax.grid()
ax.plot(X_data,y_data)
ax.set_title("True data")
ax.set_xlabel("X_data")
ax.set_ylabel("y_data");
```

```python
class TFPolynomialRegression:
    def __init__(self,
                 degrees,
                 learning_rate = 0.001,
                 epochs=100,
                 verbose=10):
        self._learning_rate = learning_rate 
        self._epochs = epochs 
        self._degrees = degrees
        self._verbose = 10
        self._errors = None
        self._W = None
        
    def fit(self,
            X_train,
            y_train):
        X=tf.placeholder(tf.float32, name="X_data")
        Y=tf.placeholder(tf.float32, name="Y_data")
        W_s = []
        y_pred = tf.Variable(tf.random.normal([1]),"y_pred")
        for power_degree in range(1,self._degrees+1):
            W = tf.Variable(tf.random.normal([1]),
                            name=f"W_{power_degree}")
            y_pred = tf.add(tf.multiply(tf.pow(X,power_degree),W),
                            y_pred)
            W_s.append(W)


        cost = tf.reduce_sum(tf.square(Y-y_pred))

        optimizer = tf.train.GradientDescentOptimizer(
                self._learning_rate)\
            .minimize(cost)
        
        self._errors = []
        
        init = tf.global_variables_initializer()
        with tf.Session() as sess:
            sess.run(init)
            for i in range(self._epochs):
                sess.run(optimizer, feed_dict={X:X_train, Y:y_train})
                loss = sess.run(cost, feed_dict={X:X_train, Y:y_train})
                self._errors.append(loss)
                if self._verbose:
                    if i%10==0:
                        logging.info(f"{i} Iteration: loss {loss}")     
            W_vals = np.array([sess.run(W) for W in W_s])  
        self._W = W_vals
        self._plot_error()
        return self
    
    def _plot_error(self):
        fig,ax = plt.subplots()
        ax.plot(np.array(self._errors))
        ax.grid()
        ax.set_title(f"Loss")
        fig.show()
    
    def predict(self,X_test):
        y = np.zeros(X_test.shape[0])
        for degree in range(self._degrees):
            y += self._W[degree]*X_test**(1+degree)
        return y
    
```

```python
polyreg = TFPolynomialRegression(degrees=4,epochs=100)
```

```python
polyreg.fit(X_data,y_data)
```

```python
y_pred = polyreg.predict(np.arange(-2,2))
```

```python
fig,ax = plt.subplots()
ax.plot(X_data,y_data,label="true")
ax.plot(np.arange(-2,2),y_pred,label="pred")
ax.grid()
ax.set_title("Data")
ax.legend()
fig.show()
```
